```python
import torch

X = torch.tensor([1,2,3,4,5], dtype = torch.float32)
Y = torch.tensor([2,4,6,8,11], dtype = torch.float32)

# we need the 'w' derivative with respect to loss function
w = torch.tensor(0.0, dtype=torch.float32, requires_grad=True)

def forward(x):
  return w * x

def loss(y,y_pred):
  return ((y - y_pred)**2).mean()

iters = 1000
learning_rate = 0.01

for epoch in range(iters):
#  print(w)
  y_pred = forward(X)
#  print(y_pred)
  l = loss(Y, y_pred)
#  print(l)
  # for the gradient of loss w/ respect to 'w'
  l.backward() # dl/dw

  with torch.no_grad():
    w.copy_(w - learning_rate * w.grad)
    w.grad.zero_()
  print(f" epoch {epoch+1:.8f} w: {w:.3f} l: {l:.8f}")
```
